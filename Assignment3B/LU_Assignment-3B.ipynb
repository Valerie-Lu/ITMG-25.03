````python
## Code goes here
import pandas as pd

population_df = pd.read_csv("recweekdata.txt",skiprows=0)
````

````python
population_df
````

````python
profile_df = pd.read_csv("studentprofiles.txt",skiprows=0)
````

````python
profile_df
````

````python
population_df = population_df.merge(profile_df, left_on =['firstname','lastname'], right_on=['firstname','lastname'])
````

````python
set(population_df.loc[:,"birthmonth"]) #checking for the different ways they inputted the month
````

````python
nummonths = {"1":"January","2":"February","3":"March","4":"April","5":"May","6":"June","7":"July","8":"August","9":"September","10":"October","11":"November","12":"December"}
shortcutmonth ={"Jan":"January","Feb":"February","Mar":"March","Apr":"April","May":"May","Jun":"June","Jul":"July","Aug":"August","Sep":"September","Oct":"October","Nov":"November","Dec":"December"}
normalmonths ={"January":"January","February":"February","March":"March","April":"April","June":"June","July":"July","August":"August","September":"September","October":"October","November":"November","December":"December"}
````

````python
for i in population_df.loc[:,"birthmonth"]:
    try: 
        population_df.loc[:,"birthmonth"].replace(i, nummonths[i], inplace=True)
    except KeyError:
        try:
            population_df.loc[:,"birthmonth"].replace(i, shortcutmonth[i],inplace=True)
        except KeyError:
            population_df.loc[:,"birthmonth"].replace(i, normalmonths[i],inplace=True)
````

````python
population_df
````

````python
set(population_df.loc[:,"birthyear"]) #checking for the different ways they inputted the year
````

````python
years= {"97":"1997","98":"1998","99":"1999", "1997":"1997","1998":"1998","1999":"1999","2000":"2000","2001":"2001"}
population_df["birthyear"] = population_df.loc[:,"birthyear"].astype(str).apply(lambda x: years[x])
````

````python
population_df #test if it worked
````

````python
set(population_df.loc[:,"course"])
````

````python
population_df.loc[:,"course"].replace(r".*C.*T.*", "BS Communications Technology", regex=True, inplace=True)
population_df.loc[:,"course"].replace(r".*I.*T.*E.*", "BS Information Technology Entrepreneurship", regex=True, inplace=True)
population_df.loc[:,"course"].replace(r".*L.*M.*", "BS Legal Management", regex=True, inplace=True)
population_df.loc[:,"course"].replace(r".*M.*E.*", "BS Management Engineering",regex=True, inplace=True)
population_df.loc[:,"course"].replace(r".*M.*[tT]$", "BS Management Engineering",regex=True, inplace=True)
````

````python
population_df
````

````python
population_df.loc[:,"phonenumber"].replace(r"^(9.+)", r"+63 \1",regex=True, inplace=True)
````

````python
carriers = {"+63 917":"Globe","+63 918":"Smart", "+63 927":"Globe"}
````

````python
population_df["carriers"] = population_df.loc[:,"phonenumber"].str[0:7].apply(lambda x: carriers[x])
````

````python
population_df
````

````python
#make clean csv

population_df.to_csv("cleanrecweekdata3b.csv")
````

````python
population_df.groupby(["birthmonth"])["phonenumber"].count()
````

````python
sortedmonthlist = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]
sortedmonthtally ={}

for i in sortedmonthlist:
    sortedmonthtally[i]= (population_df.groupby(["birthmonth"])["phonenumber"].count()).loc[i]
    
sortedmonthtally
````

````python
html_birthmonth = '''<div class="table1">
<h2>1. How many people were born in each month?</h2>
<div class='rows'>
'''
for i in sortedmonthtally:
    html_birthmonth += "<div class='cells'>{}</div>".format(i)
html_birthmonth+= "</div>"

html_birthmonth+= "<div class ='rows'>"
for i in sortedmonthtally:
    html_birthmonth += "<div class='cells'>{}</div>".format(sortedmonthtally[i])
html_birthmonth+= "</div>"
html_birthmonth+= "</div>"
with open('html_birthmonth.html',"w+") as file:
    file.write(html_birthmonth)
````
