````python
## Code goes here
import re
import csv 

dirtyrecweek = dict()

with open('recweekdata.txt') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    i = 0
    for row in csv_reader:
        if line_count == 0:
            print(f'Column names are {", ".join(row)}') 
        else:
            dirtyrecweek[i]={"firstname":row[0],"lastname":row[1], "birthmonth":row[2], "birthday":row[3], "birthyear":row[4],"course":row[5],"phonenumber":row[6] }
        i += 1
        line_count += 1

dirtyrecweek
````

## Cleaning Phone Numbers

````python
phonenumberslist= []

for i in range(1,3001):
    phonenumberslist.append(dirtyrecweek[i]["phonenumber"])
    
phonenumberslist
````

````python
newphonenumberslist =[]
for i in range(0,3000):
    number = phonenumberslist[i]
    new_number = re.sub(r"^(9[0-9][0-9])",r"+63 \1",number)
    newphonenumberslist.append(new_number)
newphonenumberslist
````

````python
counterr = 0
unique= set(newphonenumberslist)
for number in unique:
    counterr += 1

counterr
````

````python
carierlist = []
cleancariertally={"Globe":0, "Smart":0}

for number in newphonenumberslist: #substitute numbers for 'Globe' and 'Smart'
    match = re.search(r'[+]63\s917\s[0-9]{3}\s[0-9]{4}', number)
    if match:
        cleancariertally["Globe"]+=1
    match2 = re.search(r'[+]63\s927\s[0-9]{3}\s[0-9]{4}', number)
    if match2:
        cleancariertally["Globe"]+=1
    match3 = re.search(r'[+]63\s918\s[0-9]{3}\s[0-9]{4}', number)
    if match3:
        cleancariertally["Smart"]+=1

cleancariertally
````

## Cleaning the list
